```python
import os
from os import listdir
from os.path import join
import re
from nltk.tokenize import word_tokenize, sent_tokenize
import math
import glob
import pandas as pd
```

## Terms

```python
tijding = ['tijding', 'tyding', 'tydinge', 'tijdingh', 'tydingen', 'tydinghen', 'tijdinghe', 'tijdinghen', 'tijdinge', 'tydynghe', 'tijdijnge', 'tijdijnghe']
courant = ['courant', 'courante', 'kourant', 'courantier', 'couranten', 'krant', 'kranten']
gazette = ['gazette', 'gaset', 'gasetten', 'gasette']
nieuws = ['nieuwsbladen','nieuwsblaederen', 'nieuwsblad', 'nieuwsblaed', 'nieuws blaed', 'nieuws blaederen', 'nieuwsbladt', 'nieuws bladt', 'nieuwstijdingen', 'nieuwstijding', 'nieuwstyding''nieuwstydinge', 'nieuwstijdingh', 'nieuwstydingen', 'nieuwstydinghen', 'nieuwstijdinghe', 'nieuwstijdinghen', 'nieuwstijdinge', 'nieuwstydynghe', 'nieuwstijdijnge', 'nieuwstijdijnghe']
advertentie = ['advertentie', 'advertentien', 'advertenties']


all_terms = []
all_terms.extend( tijding )
all_terms.extend( courant )
all_terms.extend( gazette )
all_terms.extend( nieuws )

print( f'{len(all_terms)} terms.' )
```

```python

def concordance( text, regex , width = 10 ):

    concordance = []
    distance = math.floor( width /2 )
    regex = re.sub('[*|+]' , '' , regex )

    segment_length = 0

    words = word_tokenize( text )
    words = remove_punctuation( words )
    i = 0
    for w in words:
        if re.search( regex , str(w) , re.IGNORECASE ):
            match = ''
            for x in range( i - distance , ( i + distance ) + 1 ):
                if x >= 0 and x < len(words):
                    if len(words[x]) >= 0:
                        match += words[x] + ' '
    
            concordance.append( match.strip() )

        i += 1

    return concordance



def remove_punctuation(text):
    punctuation = "[{`,.?!:;/\()''""Â¬}]"

    new_text = ''
    for c in text:
        if c not in punctuation:
            new_text += c

    return new_text

def count_occurrences(text,terms):
    words = word_tokenize(text)

    freq = dict()
    for w in words:
        if w.lower() in terms: 
            freq[w.lower()] = freq.get(w.lower(),0)+1
    return freq
            
```

```python
path = glob.glob(os.path.join('corpus_220915/*.txt'))

rows = []

for i,file in enumerate(path):
    print( f'{file} [{i}/{len(path)}]')
    
    file_id = re.search(r"(?<=915/)(.*?)(?=.txt)", file).group(0)
    year = file_id[:4]
    
    full_text = ''
    with open( file , encoding = 'utf-8' ) as f:
        sections = f.readlines()
        full_text = ' '.join(sections)
    full_text = re.sub( r'\s+' , ' ' , full_text.strip() )
    full_text = re.sub( r'[-]\s', '' , full_text )
    for s in sent_tokenize(full_text): 
        if re.search( r';\s+(?=[A-Z])' , s ):
            sub_clauses = re.split( r'[;:]\s+(?=[A-Za-z])' , s )
            for clause in sub_clauses:
                freq = count_occurrences(clause,all_terms)
                if sum(freq.values()) > 0:
                    rows.append( [file_id,year,list(freq.keys()),remove_punctuation(clause)] )

        else:
            freq = count_occurrences(s,all_terms)
            if sum(freq.values()) > 0:
                if len(s) > 250:
                    for term in freq:
                        c = concordance( s , term , 24 )
                        for fragment in c:
                            rows.append( [file_id,year,list(freq.keys()),remove_punctuation(fragment)] )
                else:
                    rows.append( [file_id,year,list(freq.keys()),remove_punctuation(s)] )

```

```python
df = pd.DataFrame( rows , columns = ['file_id','year','found_terms','fragment'] )
```

```python
df.to_csv('newspapers.csv' , index= False )
```
